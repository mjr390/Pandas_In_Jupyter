```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.arange(0, 130, 10)
dangerD = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
danger_drop, = plt.plot(time, dangerD, color="red", label="Danger Drop")
railgun, = plt.plot(time, rail, color="blue", label="Railgun")
plt.grid()
```

```python
plt.xlim(0,120)
plt.ylim(5,95)
```

```python
plt.legend(handles=[danger_drop, railgun], loc="best")
```
