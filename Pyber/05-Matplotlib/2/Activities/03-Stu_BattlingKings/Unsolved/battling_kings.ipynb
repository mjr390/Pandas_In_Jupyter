```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
useFile= pd.read_csv("Resources/got.csv")
useFile
```

```python
# Get attacker and defender data
attacker= useFile["attacker_king"].value_counts()
defender= useFile["defender_king"].value_counts()
```

```python
# Get total battle data
total_battle= attacker.add(defender, fill_value=0)
total_battle
```

```python
# Configure plot and ticks
total_battle.plot(kind="bar", facecolor="r", title="Got Battles", )
```

```python
# Set textual properties
plt.xlabel("X Labels")
plt.ylabel("TEST")
```
